```
import random

random.seed(42)
x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]

colors = ["red", "green", "blue", "yellow", "pink",
          "orange", "purple", "brown", "cyan", "magenta"]

plt.scatter(x, y, c=colors)
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.show()
```

```
years    = [2017, 2018, 2019, 2020, 2021, 2022]
bears    = [8,   54,  93,  116, 137, 184]
dolphins = [150, 77,  32,  11,  6,   1  ]
whales   = [80,  54,  100, 76,  93,  72 ]
animals  = ['Bears', 'Dolphins', 'Whales']


for i, year in enumerate(years):
    data = [bears[i], dolphins[i], whales[i]]
    plt.pie(data, labels=animals, autopct='%1.1f%%')
    plt.title(f"Animal Population {year}")
    plt.show()


fig, axes = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle("Animal Population 2017-2022", fontsize=14)

for idx, year in enumerate(years):
    ax   = axes[idx // 3][idx % 3]
    data = [bears[idx], dolphins[idx], whales[idx]]
    ax.pie(data, labels=animals, autopct='%1.1f%%')
    ax.set_title(str(year))

plt.tight_layout()
plt.show()
```

```
car_colors  = ['Red', 'Silver', 'Blue', 'White', 'Green']
cars_seen   = [123, 230, 78, 193, 12]

plt.bar(car_colors, cars_seen)
plt.title("Cars Seen by Colour")
plt.xlabel("Car Colour")
plt.ylabel("Number Seen")
plt.show()

plt.bar(car_colors, cars_seen, width=0.3)
plt.title("Cars Seen by Colour (narrow bars)")
plt.xlabel("Car Colour")
plt.ylabel("Number Seen")
plt.show()

plt.barh(car_colors, cars_seen)
plt.title("Cars Seen by Colour (horizontal)")
plt.xlabel("Number Seen")
plt.ylabel("Car Colour")
plt.show()

bar_fill_colors = ['red', 'silver', 'blue', 'white', 'green']
plt.bar(car_colors, cars_seen, color=bar_fill_colors, edgecolor='black')
plt.title("Cars Seen by Colour (matched colours)")
plt.xlabel("Car Colour")
plt.ylabel("Number Seen")
plt.show()

plt.pie(cars_seen, labels=car_colors, autopct='%1.1f%%')
plt.title("Cars Seen by Colour")
plt.show()

myexplode = [0.3, 0, 0, 0, 0]
plt.pie(cars_seen, labels=car_colors, explode=myexplode, autopct='%1.1f%%')
plt.title("Cars Seen – Exploded Red slice")
plt.show()

plt.pie(cars_seen, labels=car_colors, explode=myexplode,
        shadow=True, autopct='%1.1f%%')
plt.title("Cars Seen – Exploded + Shadow")
plt.show()


plt.pie(cars_seen, labels=car_colors, autopct='%1.1f%%')
plt.legend(title="Car Colours:")
plt.title("Cars Seen by Colour")
plt.show()


mycolours = ["Magenta", "Green", "Cyan", "Red", "Orange"]
plt.pie(cars_seen, labels=car_colors, colors=mycolours, autopct='%1.1f%%')
plt.title("Cars Seen – Custom Colours")
plt.show()
```
