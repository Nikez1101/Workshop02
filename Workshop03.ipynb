```python
x_range = list(range(0, 20))
y_f1 = [5*x**3 + 2*x - 1      for x in x_range]
y_f2 = [-2*x**3 + x**2 + 100  for x in x_range]
y_f3 = [2*math.pi*x + 20      for x in x_range]

plt.figure(figsize=(14, 4))

plt.subplot(1, 3, 1)
plt.plot(x_range, y_f1, marker='o', color='blue')
plt.title('f(x) = 5x³+2x−1')
plt.xlabel('X'); plt.ylabel('Y'); plt.grid()

plt.subplot(1, 3, 2)
plt.plot(x_range, y_f2, marker='s', color='green')
plt.title('f(x) = −2x³+x²+100')
plt.xlabel('X'); plt.ylabel('Y'); plt.grid()

plt.subplot(1, 3, 3)
plt.plot(x_range, y_f3, marker='^', color='red')
plt.title('f(x) = 2πx+20')
plt.xlabel('X'); plt.ylabel('Y'); plt.grid()

plt.tight_layout()
plt.show()
```

```python
x = []
y = []

start = -3 * math.pi
end   =  3 * math.pi
step  = 0.1

i = start
while i <= end:
    x.append(i)
    y.append(math.cos(i))
    i += step

plt.plot(x, y, color='blue', label='cos(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y = cos(x),  −3π ≤ x ≤ 3π')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = []
y = []

r = 4
a, b = 2, 2

theta = 0.0
theta_end = 2 * math.pi + 0.01
step_size = 0.05

while theta <= theta_end:
    x.append(r * math.cos(theta) + a)
    y.append(r * math.sin(theta) + b)
    theta += step_size

plt.figure()
plt.plot(a, b, '.', markersize=8, color='black', label='Center (2,2)')
plt.plot(x, y, '-', color='blue', label='Circle r=4')
plt.xlabel('X'); plt.ylabel('Y')
plt.title('Circle: center (2,2), radius 4')
plt.legend()
plt.grid()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python

xc, yc = [], []
r = 4; cx, cy = 2, 2
theta = 0.0
while theta <= 2*math.pi + 0.01:
    xc.append(r * math.cos(theta) + cx)
    yc.append(r * math.sin(theta) + cy)
    theta += 0.05


xe, ye = [], []
a_e = 3; b_e = 5; p, q = 2, 2
theta = 0.0
while theta <= 2*math.pi + 0.01:
    xe.append(a_e * math.cos(theta) + p)
    ye.append(b_e * math.sin(theta) + q)
    theta += 0.05

plt.figure()
plt.plot(cx, cy, 'k.', markersize=8, label='Center (2,2)')
plt.plot(xc, yc, '-b', label='Circle (r=4)')
plt.plot(xe, ye, '-r', label='Ellipse (a=3, b=5)')
plt.xlabel('X'); plt.ylabel('Y')
plt.title('Circle and Ellipse on Same Axes')
plt.legend()
plt.grid()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
p = 0
q = 0
a = 0.5


x_values = []
y_values = []

step = 0.1
y = -5.0
while y <= 5.0:
    x = (y - p)**2 / (4 * a) + q
    x_values.append(x)
    y_values.append(y)
    y += step

plt.figure()
plt.plot(x_values, y_values, '-g', linewidth=2, label=f'Parabola vertex=({q},{p}), a={a}')
plt.plot(q, p, 'ro', markersize=8, label='Vertex')
plt.xlabel('X'); plt.ylabel('Y')
plt.title(f'Parabola: (y−{p})² = 4·{a}·(x−{q})')
plt.legend()
plt.grid()
plt.show()
```
