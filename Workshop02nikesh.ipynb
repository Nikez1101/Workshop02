```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

n = int(input("Enter a number: "))
if is_prime(n):
    print(f"{n} is Prime")
else:
    print(f"{n} is Not Prime")
```

```python
lst = (input("Enter a list: "))
unique_list = []

for item in lst:
    if item not in unique_list:
        unique_list.append(item)

print("Unique List :", unique_list)
```

```python
n = int(input("Enter a number: "))
binary = ""

while n > 0:
    binary = str(n % 2) + binary
    n = n // 2

print("Binary:", binary)
```

```python
def prodDigits():
    n = int(input("Enter a number: "))
    n = abs(n)
    product = 1
    for digit in str(n):
        product *= int(digit)
    return product
prodDigits()
```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

def findNextPerfect():
    n = int(input("Enter a number: "))
    while True:
        if sumOfDivisors(n) == n:
            print(f"{n} is next perfect number")
            return n
        n += 1
findNextPerfect()
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
mean = sum(marks) / len(marks)

# Median
sorted_marks = sorted(marks)
n = len(sorted_marks)
mid = n // 2

if n % 2 == 0:
    median = (sorted_marks[mid - 1] + sorted_marks[mid]) / 2
else:
    median = sorted_marks[mid]

print("Mean:", mean)
print("Median:", median)
```
